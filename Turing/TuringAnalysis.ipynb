```python
# Which country has the highest death rate
ps.sqldf('''
    SELECT 
        a.location, 
        total_cases/population as death_rate,
        (CASE
            WHEN total_cases/population > 0.00005 THEN 1
            ELSE 0
        END) AS matched
    FROM
        (SELECT 
            location,
           SUM(new_deaths) as total_cases
        FROM 
            covid_data
        WHERE 
            aged_65_older_percent > 20
        GROUP BY 
            location) a
        JOIN
            covid_data_pop b ON (a.location = b.location)
    ORDER BY 
        death_rate DESC
 ''')
```
